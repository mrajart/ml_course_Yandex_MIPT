## Cравнение предложений

```python
file = open("sentences.txt")
```

```python
text = file.readlines()
```

```python
text
```

```python
lines = []
for line in text :
    lines.append(line.lower())
```

```python
import re
```

```python
sents = []
wordstemp = []
for line in lines:
    wordstemp = re.split('[^a-z]', line)
    temp = []
    for word in wordstemp:
        if word != '':
            temp.append(word)
    sents.append(temp)
```

```python
sents
```

```python
allwords = []
for line in sents:
    for word in line:
        allwords.append(word)
print(allwords)
```

```python
dictwords={}
numb=0
for word in allwords:
    if word not in dictwords:
        dictwords[word] = numb
        numb += 1
print(dictwords)
```

```python
import numpy as np
```

```python
d = numb
n = len(sents)
matrix = np.zeros((n, d))
matrix.shape
```

```python
'''for word in allwords:
    countin = 0
    nline = 0
    for line in sents:
        for wordLine in line : 
            if word == wordLine:
                countin +=1
        matrix[nline][dictwords[word]] = countin
        nline += 1'''
for i in range(n):
    tokens = sents[i]
    for token in tokens:
        matrix[i][dictwords[token]] += 1
```

```python
matrix
```

```python
from scipy.spatial import distance
```

```python
dist = [] 
u = matrix[0,] 
for i in range(1, len(sents)):
    v = matrix[i,]
    dist.append({"index": i, "distance": distance.cosine(u, v)}) 
```
